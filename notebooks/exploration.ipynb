```python
#Testing embeddings model
model=SentenceTransformer('all-MiniLM-L6-v2')
test_text="This text is for test purpose"
embeddings=model.encode(test_text)
print(f" Embeddings shape:{embeddings.shape}")
```

```python
#Testing chromaDB
client=chromadb.Client()
collection = client.create_collection("test")
```

```python
# Provide precomputed embeddings to avoid the client's remote embedding call which may time out.
# `embeddings` variable is already computed in another cell.
collection.add(
    documents=["This is a test document"],
    ids=["test1"],
    embeddings=[embeddings.tolist()]
)
print("ChromaDB working")
```
